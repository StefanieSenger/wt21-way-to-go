## Merging raw json data files into a single data frame and writing to .csv

```python
import pandas as pd
import json
import glob
```

```python
#this code assumes you have the raw json data in a directory titled 'poi_data', which should be in the same directory as this file
all_node_files = glob.glob('poi_data/' + '/*node.json')

li = []

for filename in all_node_files:
    data = json.load(open(filename))
    df = pd.DataFrame(data["elements"])
    li.append(df)

all_nodes = pd.concat(li, axis=0, ignore_index=True)
```

```python
all_nodes
```

```python
all_nodes.to_csv('all_nodes_uncleaned.csv')
```
