# Clustering

```python
import matplotlib.pyplot as plt
from kneed import KneeLocator
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.preprocessing import StandardScaler
```

```python
kmeans = KMeans(
    init="random",
    n_clusters=3,
    n_init=10,
    max_iter=300,)

kmeans.fit(y)
```

```python
# The lowest SSE value
kmeans.inertia_
```

```python
# Final locations of the centroid
kmeans.cluster_centers_
```

```python
# The number of iterations required to converge
kmeans.n_iter_
```

```python
kmeans = KMeans(
    init="random",
    n_clusters=3,
    n_init=50,
    max_iter=500,)

kmeans.fit(y)
```

```python
print('The lowest SSE value', kmeans.inertia_)
print('Final locations of the centroid', kmeans.cluster_centers_)
print('The number of iterations required to converge', kmeans.n_iter_)
```

```python
label = kmeans.labels_
label
```

```python
# Elbow method

kmeans_kwargs = {
    "init": "random",
    "n_init": 12,
    "max_iter": 500,
}

# A list holds the SSE values for each k
sse = []
for k in range(1, 12):
    kmeans = KMeans(n_clusters=k, **kmeans_kwargs)
    kmeans.fit(y)
    sse.append(kmeans.inertia_)
    
# Plotting
plt.style.use("fivethirtyeight")
plt.plot(range(1, 12), sse)
plt.xticks(range(1, 12))
plt.xlabel("Number of Clusters")
plt.ylabel("SSE")
plt.show()
```

```python
# Determining the elbow point in the SSE 

kl = KneeLocator(
    range(1, 12), sse, curve="convex", direction="decreasing"
)

kl.elbow
```

```python
# Silhouette coefficient

# A list holds the silhouette coefficients for each k
silhouette_coefficients = []

# Notice you start at 2 clusters for silhouette coefficient
for k in range(2, 12):
    kmeans = KMeans(n_clusters=k, **kmeans_kwargs)
    kmeans.fit(y)
    score = silhouette_score(y, kmeans.labels_)
    silhouette_coefficients.append(score)

# Plotting
plt.style.use("fivethirtyeight")
plt.plot(range(2, 12), silhouette_coefficients)
plt.xticks(range(2, 12))
plt.xlabel("Number of Clusters")
plt.ylabel("Silhouette Coefficient")
plt.show()
```

## Now try with the Pipeline

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import LabelEncoder, MinMaxScaler

n_clusters = len(label)

preprocessor = Pipeline(
    [
        ("scaler", MinMaxScaler()),
        ("pca", PCA(n_components=3)),
    ]
)

clusterer = Pipeline(
   [
       (
           "kmeans",
           KMeans(
               n_clusters=n_clusters,
               init="k-means++",
               n_init=50,
               max_iter=500,
           ),
       ),
   ]
)

pipe = Pipeline(
    [
        ("preprocessor", preprocessor),
        ("clusterer", clusterer)
    ]
)
```

```python
pipe.fit(df)
```

```python
preprocessed_data = pipe["preprocessor"].transform(df)

#predicted_labels = pipe["clusterer"]["kmeans"].label_

silhouette_score(preprocessed_data, label)
```

```python
pcadf = pd.DataFrame(
    pipe["preprocessor"].transform(df),
    columns=["component_1", "component_2", "component_3"],
)

pcadf["predicted_cluster"] = pipe["clusterer"]["kmeans"].labels_

plt.style.use("fivethirtyeight")
plt.figure(figsize=(8, 8))

scat = sns.scatterplot(
    "component_1",
    "component_2",
    "component_3",
    s=50,
    data=pcadf,
    hue="predicted_cluster",
    palette="Set2",
)

scat.set_title(
    "Clustering results from PCA"
)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.0)

plt.show()
```

```python
# Split data into 3
import turicreate as tc

# Load data
data = tc.SFrame('clustered_data.csv')

# Turicreate has a library named as random split that will the data randomly among the train,test. Dev will be part of test set and we will split that data later.
train_data_set,test_data=data.random_split(.6, seed=0)

# In this 0.6 it means that we will have 60% as our training data and rest 40% data as test data.
# Here seed is for giving the same set for train and test again and again.
# Now we will split our test_data into two different sets of equal length:
test_data_set,dev_set=test_data.random_split(.5, seed=0)
```
