# Example Routes for the WD Team

```python
import sqlite3
import pandas as pd

con = sqlite3.connect('routes.db') 
cur = con.cursor()
                 
cur.execute('''SELECT * FROM routes ORDER BY route_id''')

df = pd.read_sql_query('SELECT * FROM routes ORDER BY route_id;', con)

print(df)
```
